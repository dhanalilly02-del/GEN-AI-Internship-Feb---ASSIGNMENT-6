1:Smart Parking Lot Management System

```python
def smart_parking_management(capacity, logs):
    current_parked = 0
    peak_usage = 0
    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            current_parked -= 1
        else:
            print(f"Invalid log entry ignored: {log}")
        if current_parked < 0:
            current_parked = 0
        if current_parked > peak_usage:
            peak_usage = current_parked
    if current_parked > capacity:
        status = "Over Capacity"
    else:
        status = "Available"
    print(f"Currently Parked Vehicles: {current_parked}")
    print(f"Peak Parking Usage: {peak_usage}")
    print(f"Parking Status: {status}")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_management(capacity, logs)
```

2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_speed = 30  # km per hour
    base_time = (distance / base_speed) * 60
    if traffic == "Low":
        base_time *= 1.0
    elif traffic == "Medium":
        base_time *= 1.2
    elif traffic == "High":
        base_time *= 1.5
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    return round(base_time)
distance = 8
traffic = "High"
weather = "Rainy"
eta = estimate_delivery_time(distance, traffic, weather)
print(f"Estimated Delivery Time: {eta} minutes")
```

3:Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"
    print(f"Occupancy: {round(occupancy)}%")
    print(f"Show Status: {status}")
total_seats = 200
booked_seats = [1] * 150
analyze_seat_occupancy(total_seats, booked_seats)
```

 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average = sum(cpu_readings) / len(cpu_readings)
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print(f"Average CPU Load: {round(average)}%")
    print(f"Server Status: {status}")
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

5:Smart Classroom Resource Usage Monitor

```python
def smart_classroom_monitor(resource_usage):
    limits = {
        "Projector": 5,
        "AC": 8,
        "Lights": 5
    }
    overused = []
    for resource, hours in resource_usage.items():
        if resource in limits and hours > limits[resource]:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
smart_classroom_monitor(usage)
```

6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
capacity = 100
registrations = 105
event_registration_controller(capacity, registrations)
```
